```python
#importing our libraries

# import our pandas library
import pandas as pd

# import our numpy library
import numpy as np
```

```python
# loading our csv dataset

url = '/content/Autolib_dataset.csv'
df= pd.read_csv(url)
df.head(10)
```

```python
# dropping columns we are not going to use in our analysis

df.drop(['Displayed comment', 'Scheduled at','Public name', 'Rental status',	'Slots', 'Status', 'Subscription status', 
         'Charge Slots','Charging Status', 'ID', 'Kind'], axis=1, inplace=True)
```

```python
# removing any white spaces in the columns 

df.columns=df.columns.str.strip()
```

```python
# looking for any missing values

df.isnull().sum()
```

```python
# checking for any duplicates

df.duplicated()
```

```python
# checking the number of rows and columns before dropping the duplicates

df.shape
```

```python
# if any duplicate, lets drop them

df.drop_duplicates()
```

```python
# checking the number of rows and columns after dropping the duplicates

df.shape
```

```python
# turns out there were no duplicates since there is no change in the number of rows and columns
```

```python
# for uniformity in our table we are going to use the string replace function

df.columns = df.columns.str.replace(' ', '_').str.lower()
df.head(5)
```

```python
# before identifying the most popular hour of the day for picking up a bluecar,
# we are going to group the bluecar column with the city, year, month and hour column

new_df=df[['bluecar_counter', 'city', 'year', 'month', 'hour']]
new_df
```

```python
# after that we specify the values in each column 

new_df=new_df[(new_df['city']=='Paris') & (new_df['month']==4) & (new_df['year']==2018)]
new_df
```

```python
# finding the median to have a central value to work with when answering the questions

df['bluecar_counter'].median()
```

```python
#Identifying the most popular hour of the day for picking up a shared bluecar

# since the bluecar_counter column represents the number of cars available at the station, 
# we will have to look at column values with a lower number of cars to answer our
# question of finding the most popular hour

hour=new_df[(new_df['bluecar_counter'] <= 1)]
hour
```

```python
# now to find out the hour that appears the most during periods where number of bluecars
# are at their lowest
# we then find out the most popular hour of the day for picking up a shared bluecar is 21 hours

hour['hour'].mode()
```

```python
# the most popular hour for returning cars
# we are going to look at the hour where the number of bluecars start to increase
# I'll start from range 3 

hour=new_df[(new_df['bluecar_counter']>=3)]
hour
```

```python
# now finding the hour that appears the most 

hour['hour'].mode()

# we get the most popular hour for returning cars is 6 hours
```

```python
# finding the most popular station
# I am going to use the geo_point column which represents the GPS location of the station

df['geo_point'].mode()
```

```python
# finding the most popular station at the most popular picking hour

station=df[(df['hour']==21) & (df['geo_point'])]
station
station['geo_point'].mode()
```

```python
# the most popular postal code for picking up bluecars 

postal_code=df[['postal_code', 'bluecar_counter']]
postal_code
postal_code['postal_code'].mode()
```

```python
# the most popular postal code for picking up bluecars at the most popular picking hour

postal_code2=df[['postal_code', 'bluecar_counter', 'hour']]
postal_code2
postal_code2=postal_code2[(postal_code2['hour']==21)]
postal_code2['postal_code'].mode()
```
